# Bivariate Detection:

Examining Relationship between Variables

When we are finding relationship between two numeric variables we can find it by using scatterplots.
From scatter plots if we observe some cases that fall markedly outside the range of other observations will be seen as isolated points in scatterplots

```python
import pandas as pd
import numpy as np
dataset=pd.read_csv('adult.csv')
dataset
```

```python
dataset.head()
```

```python
# plotting scatter plot

dataset.plot.scatter(x='age', y='occupation')
```

Observation : age and occupation column are highly correlated 

```python
dataset.plot.scatter(x='hours-per-week', y='occupation')
```

Observation : data hours-per-week and occupation ,that means which type of occcupation needs how much hour to do the work . 
    these data are ploted. 

```python
# pair plot 
import seaborn as sns
iris=sns.load_dataset('iris')
sns.pairplot(iris)
```

```python
penguins=sns.load_dataset('penguins')
sns.pairplot(penguins)
```

```python
sns.pairplot(penguins,hue='species')
```

```python
import matplotlib.pyplot as plt
dataset=sns.load_dataset('tips')
sns.pairplot(dataset,hue='day')
plt.show()
```

```python
# barplot
import seaborn as sns
sns.barplot(x='occupation', y='age', data=dataset)
```

Observation : Observation : created barplot that represents occupation on x-axis and age on y-axis.
